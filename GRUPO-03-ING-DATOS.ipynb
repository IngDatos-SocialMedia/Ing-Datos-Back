### 2.3.1. Missing Data <a class="anchor" id="MissingData"></a>
In addition to incorrect datatypes, another common problem when dealing with real-world data is missing values. These can arise for many reasons and have to be either filled in or removed before we train a machine learning model. First, lets get a sense of how many missing values are in each column 

While we always want to be careful about removing information, if a column has a high percentage of missing values, then it probably will not be useful to our model. The threshold for removing columns should depend on the problem

Integridad

```python
def missing_values_table(df):
    # Calcular valores nulos
    mis_val = df.isnull().sum()
    # Calcular el porcentaje de valores nulos
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    # Combinar en un solo DataFrame
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    # Renombrar columnas
    mis_val_table_ren_columns = mis_val_table.rename(
        columns={0: 'Missing Values', 1: '% of Total Values'}
    )
    # Filtrar solo columnas con valores nulos
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns['Missing Values'] != 0
    ].sort_values('% of Total Values', ascending=False).round(1)

    print(
        "Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
        "There are " + str(mis_val_table_ren_columns.shape[0]) + 
        " columns that have missing values."
    )
    
    return mis_val_table_ren_columns
```

```python
import pandas as pd

# Definir los DataFrames y las columnas relevantes para verificar precios
dataframes = {
    "Binance": {"df": df1, "price_column": "price"},
    "CoinMarket": {"df": df2, "price_column": "Price (USD)"},
    "GeckoCoin": {"df": df3, "price_column": "current_price"},
    "CryptoCompare": {"df": df4, "price_column": "price"},
    "CoinLayer": {"df": df5, "price_column": "price"}
}

# Umbral para considerar precios cercanos a 0
threshold = 1e-6

# Verificar precios en cada DataFrame
for name, info in dataframes.items():
    df = info["df"]
    price_column = info["price_column"]

    print(f"--- Verificacin en {name} ---")
    
    # Asegurarse de que la columna de precios sea numrica
    try:
        df[price_column] = pd.to_numeric(df[price_column], errors='coerce')
        # Filtrar filas con precios iguales a 0 o muy cercanos a 0
        invalid_prices = df[(df[price_column] == 0) | (df[price_column] < threshold)]
        
        if not invalid_prices.empty:
            print(f"Se encontraron {len(invalid_prices)} filas con precios iguales o cercanos a 0:")
            print(invalid_prices)
        else:
            print("No se encontraron precios iguales o cercanos a 0.")
    except Exception as e:
        print(f"No se pudo procesar la columna '{price_column}' en {name}: {e}")
    
    print("\n")
```

### 2.3.2. Outliers <a class="anchor" id="Outliers"></a>
At this point, we may also want to remove outliers. These can be due to typos in data entry, mistakes in units, or they could be legitimate but extreme values. For this project, we will remove anomalies based on the definition of extreme outliers:

https://www.itl.nist.gov/div898/handbook/prc/section1/prc16.htm

- Below the first quartile  3  interquartile range
- Above the third quartile + 3  interquartile range

Precision

```python
def detect_outliers_iqr(df, column):
    # Calcular el primer cuartil (Q1) y tercer cuartil (Q3)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    # Calcular el rango intercuartlico (IQR)
    IQR = Q3 - Q1
    
    # Definir los lmites inferior y superior
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identificar los outliers que estn fuera de los lmites
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    print(f"Se encontraron {len(outliers)} outliers en la columna '{column}'")
    return outliers
```

```python
dataframes = {
    "Binance": {"df": df1, "price_column": "price"},
    "CoinMarket": {"df": df2, "price_column": "Price (USD)"},
    "GeckoCoin": {"df": df3, "price_column": "current_price"},
    "CryptoCompare": {"df": df4, "price_column": "price"},
    "CoinLayer": {"df": df5, "price_column": "price"}
}

# Detectar outliers para cada DataFrame
for name, info in dataframes.items():
    df = info["df"]
    price_column = info["price_column"]
    
    print(f"--- Verificacin de outliers en {name} ---")
    
    # Verificar outliers en la columna de precios
    outliers = detect_outliers_iqr(df, price_column)
    
    # Si se encuentran outliers, se imprimen
    if not outliers.empty:
        print(f"Outliers detectados en {name}:")
        print(outliers)
    else:
        print(f"No se encontraron outliers en la columna '{price_column}' de {name}.")
    
    print("\n")
```

Limpieza de outliers

```python
# Funcin para detectar y limpiar outliers usando el IQR
def clean_outliers_iqr(df, column, remove_outliers=True):
    # Calcular el primer cuartil (Q1) y tercer cuartil (Q3)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    # Calcular el rango intercuartlico (IQR)
    IQR = Q3 - Q1
    
    # Definir los lmites inferior y superior
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Detectar outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Se encontraron {len(outliers)} outliers en la columna '{column}'")
    
    if remove_outliers:
        # Eliminar outliers
        df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
        print(f"Se han eliminado {len(df) - len(df_cleaned)} outliers en la columna '{column}'")
        return df_cleaned
    else:
        # Reemplazar outliers con la media o mediana (dependiendo de la preferencia)
        median = df[column].median()
        df[column] = df[column].apply(lambda x: median if x < lower_bound or x > upper_bound else x)
        print(f"Se han reemplazado los outliers con la mediana ({median}) en la columna '{column}'")
        return df

# Definir los DataFrames y las columnas relevantes
dataframes = {
    "Binance": {"df": df1, "price_column": "price"},
    "CoinMarket": {"df": df2, "price_column": "Price (USD)"},
    "GeckoCoin": {"df": df3, "price_column": "current_price"},
    "CryptoCompare": {"df": df4, "price_column": "price"},
    "CoinLayer": {"df": df5, "price_column": "price"}
}

# Limpiar outliers para cada DataFrame
for name, info in dataframes.items():
    df = info["df"]
    price_column = info["price_column"]
    
    print(f"--- Limpieza de outliers en {name} ---")
    
    # Limpiar outliers en la columna de precios
    cleaned_df = clean_outliers_iqr(df, price_column, remove_outliers=True)  # Cambia remove_outliers a False si deseas reemplazar los outliers
    
    # Mostrar las primeras filas del DataFrame limpio
    print(f"Primeras filas del DataFrame limpio en {name}:")
    print(cleaned_df.head())
    
    print("\n")
```

```python
import pandas as pd

def count_values_distribution(df, column):
    # Contar la cantidad de valores nicos
    count_val = df[column].value_counts()
    
    # Calcular el porcentaje de cada valor
    count_val_percent = 100 * count_val / len(df)
    
    # Crear una tabla con conteos y porcentajes
    count_val_table = pd.concat([count_val, count_val_percent.round(1)], axis=1)
    count_val_table.columns = ['Count Values', '% of Total Values']
    
    return count_val_table
```

### Differencing
Diferenciacin Especficamente, se construye una nueva serie en la que el valor en el paso
temporal actual se calcula como la diferencia entre la observacin original y la observacin en el
paso temporal anterior. valor(t) = observacin(t) - observacin(t-1)


```python
def apply_differencing(df, value_column, time_column=None):
    if time_column:
        # Ordenar por la columna temporal
        df = df.sort_values(by=time_column)
    
    # Aplicar la diferenciacin
    df[f"{value_column}_diff"] = df[value_column].diff()
    return df
```

```python
def data_quality_report(df, df_name):
    print(f"\nData Quality Report for {df_name}")
    
    # 1. Valores faltantes
    print("\n1. Valores faltantes por columna:")
    missing_values = df.isnull().sum()
    print(missing_values[missing_values > 0])  # Muestra solo las columnas con valores faltantes
    
    total_missing_values = missing_values.sum()
    print(f"\nTotal de valores faltantes: {total_missing_values}")
    
    # 2. Valores duplicados
    duplicated_rows = df.duplicated().sum()
    print(f"\n2. Total de filas duplicadas: {duplicated_rows}")
    
    # 3. Tipos de datos
    print(f"\n3. Tipos de datos por columna:\n{df.dtypes}")
    
    # 4. Estadsticas bsicas
    print("\n4. Estadsticas bsicas:")
    print(df.describe(include='all'))  # Incluye todas las columnas, incluso las no numricas
    
    # 5. Valores nicos por columna
    print("\n5. Valores nicos por columna:")
    for col in df.columns:
        print(f"{col}: {df[col].nunique()} valores nicos")
    
    # 6. Deteccin de valores negativos y cercanos a cero
    print("\n6. Deteccin de valores negativos y cercanos a cero:")
    for col in df.select_dtypes(include=['float64', 'int64']).columns:  # Solo columnas numricas
        negative_or_zero = df[(df[col] <= 0)]
        if not negative_or_zero.empty:
            print(f"Valores negativos o cero en la columna '{col}':")
            print(negative_or_zero[[col]].head())
    
    # 7. Deteccin de outliers utilizando IQR
    print("\n7. Deteccin de outliers (basada en IQR):")
    for col in df.select_dtypes(include=['float64', 'int64']).columns:  # Solo columnas numricas
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        
        if not outliers.empty:
            print(f"Se encontraron {len(outliers)} outliers en la columna '{col}':")
            print(outliers[[col]].head())
    
    # 8. Validacin de la coherencia de los datos (por ejemplo, revisar columnas especficas si es necesario)
    print("\n8. Validacin de coherencia de los datos:")
    # Aqu puedes agregar validaciones especficas de acuerdo a las reglas de tu negocio
    
    # 9. Limpiar outliers (eliminar o reemplazar valores atpicos)
    print("\n9. Limpieza de outliers:")
    for col in df.select_dtypes(include=['float64', 'int64']).columns:  # Solo columnas numricas
        # Limpiar outliers eliminando los valores fuera de los lmites
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    
    print("\n9. DataFrame despus de limpieza de outliers:")
    print(df.describe(include='all'))  # Muestra estadsticas despus de la limpieza
    
    # 10. Finalizacin y resultado
    print("\n10. Resumen final de la calidad de los datos:")
    print(f"Filas despus de limpieza de outliers: {len(df)}")
    print(f"Total de valores faltantes despus de limpieza: {df.isnull().sum().sum()}")
    print(f"Total de filas duplicadas despus de limpieza: {df.duplicated().sum()}")
```

This task includes constructive data preparation operations such as the production of derived attributes or entire new records, or transformed values for existing attributes.

**Derived attributes** - These are new attributes that are constructed from one or more existing attributes in the same record, for example you might use the variables of length and width to calculate a new variable of area.

**Generated records** - Here you describe the creation of any completely new records. For example you might need to create records for customers who made no purchase during the past year.
```python
# Promedio mvil del precio (Rolling Average)
df_cryptocompare['rolling_price'] = df_cryptocompare.groupby('symbol', observed=False)['price'].rolling(window=3, min_periods=1).mean().reset_index(level=0, drop=True)
df_coinmarketcap['rolling_price'] = df_coinmarketcap.groupby('symbol', observed=False)['price'].rolling(window=3, min_periods=1).mean().reset_index(level=0, drop=True)
df_coinlayer['rolling_price'] = df_coinlayer.groupby('symbol', observed=False)['price'].rolling(window=3, min_periods=1).mean().reset_index(level=0, drop=True)
df_coingecko['rolling_price'] = df_coingecko.groupby('symbol', observed=False)['price'].rolling(window=3, min_periods=1).mean().reset_index(level=0, drop=True)
df_binance['rolling_price'] = df_binance.groupby('symbol', observed=False)['price'].rolling(window=3, min_periods=1).mean().reset_index(level=0, drop=True)
```

## 3.4 Integrate Data  <a class="anchor" id="Integratedata"></a>
These are methods whereby information is combined from multiple databases, tables or records to create new records or values.

**Merged data** - Merging tables refers to joining together two or more tables that have different information about the same objects. For example a retail chain might have one table with information about each stores general characteristics (e.g., floor space, type of mall), another table with summarised sales data (e.g., profit, percent change in sales from previous year), and another with information about the demographics of the surrounding area. Each of these tables contains one record for each store. These tables can be merged together into a new table with one record for each store, combining fields from the source tables.

**Aggregations** - Aggregations refers to operations in which new values are computed by summarising information from multiple records and/or tables. For example, converting a table of customer purchases where there is one record for each purchase into a new table where there is one record for each customer, with fields such as number of purchases, average purchase amount, percent of orders charged to credit card, percent of items under promotion etc.


```python
# Merge: uniendo los DataFrames usando 'symbol' como clave
merged_df = pd.merge(df_binance[['symbol', 'price', 'rolling_price', 'price_normalized']], 
                     df_coinmarketcap[['symbol', 'price', 'rolling_price', 'price_normalized']], 
                     on='symbol', how='outer', suffixes=('_binance', '_coinmarketcap'))

merged_df = pd.merge(merged_df, df_coinlayer[['symbol', 'price', 'rolling_price', 'price_normalized']], 
                     on='symbol', how='outer', suffixes=('_coinmarketcap', '_coinlayer'))

merged_df = pd.merge(merged_df, df_coingecko[['symbol', 'price', 'rolling_price', 'price_normalized']], 
                     on='symbol', how='outer', suffixes=('_coinlayer', '_coingecko'))

merged_df = pd.merge(merged_df, df_cryptocompare[['symbol', 'price_cryptocompare', 'rolling_price', 'price_normalized']], 
                     on='symbol', how='outer', suffixes=('_coingecko', '_cryptocompare'))

# Verificar las columnas despus del merge
print(merged_df.columns)
```

2. Realizar el Anlisis Exploratorio de los Datos (EDA)
Aqu definimos la funcin principal perform_eda que realiza el anlisis exploratorio de los datos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import json

# Funcin para cargar los datos desde un archivo JSON
def load_data(file_path):
    """Carga los datos desde el archivo JSON."""
    if os.path.exists(file_path) and os.path.getsize(file_path) > 0:
        with open(file_path, "r", encoding="utf-8") as f:
            return pd.DataFrame(json.load(f))
    else:
        print(f"Archivo vaco o no encontrado: {file_path}")
        return pd.DataFrame()

# Funcin para realizar el anlisis exploratorio de los datos
def perform_eda(file_path):
    """Realiza un anlisis exploratorio de los datos."""

    # Cargar los datos
    data = load_data(file_path)

    # Ver las primeras filas de los datos
    print("Primeras filas de los datos:")
    print(data.head())

    # Estadsticas descriptivas
    print("\nEstadsticas descriptivas:")
    print(data.describe())

    # Verificar los tipos de datos y la existencia de valores nulos
    print("\nTipos de datos y valores nulos:")
    print(data.info())

    # Verificar si hay valores nulos
    print("\nCantidad de valores nulos por columna:")
    print(data.isnull().sum())

    # Asegurarse de que 'price' sea numrico
    data['price'] = pd.to_numeric(data['price'], errors='coerce')

    # Eliminar filas con NaN en la columna 'price' si no se puede convertir
    data = data.dropna(subset=['price'])

    # Visualizacin de la distribucin de precios
    plt.figure(figsize=(10, 6))
    sns.histplot(data['price'], bins=30, kde=True, color='blue')
    plt.title('Distribucin de los Precios de las Criptomonedas')
    plt.xlabel('Precio')
    plt.ylabel('Frecuencia')
    plt.show()

    # Grfico de tendencias de precios a lo largo del tiempo
    plt.figure(figsize=(10, 6))
    data['timestamp'] = pd.to_datetime(data['timestamp'])
    plt.plot(data['timestamp'], data['price'], label='Precio', color='green')
    plt.title('Tendencia de Precios de las Criptomonedas')
    plt.xlabel('Fecha')
    plt.ylabel('Precio')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()

    # Eliminar columnas no numricas para calcular la correlacin
    numeric_data = data.select_dtypes(include=['float64', 'int64'])  # Solo columnas numricas

    # Correlacin entre las variables numricas
    print("\nMatriz de correlacin:")
    correlation_matrix = numeric_data.corr()  # Calcular correlaciones entre las variables numricas
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Matriz de Correlacin')
    plt.show()

    # Identificacin de outliers (valores atpicos)
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data['price'])
    plt.title('Identificacin de Outliers en los Precios')
    plt.xlabel('Precio')
    plt.show()
```

```python
# Cargar archivo de datos
file_path = "crypto_etl_project/data/binance/binance_data_new.json"  # Ruta del archivo de datos actualizado

# Ejecutar el anlisis exploratorio de los datos
perform_eda(file_path)
```
